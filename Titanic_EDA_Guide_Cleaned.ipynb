```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import seaborn as sns
sns.set(style="whitegrid")

# Load data
df = pd.read_csv("Titanic-Dataset.csv")
df.head()
```

```python
# Summary statistics
df.describe()
```

```python
# Age Histogram and Boxplot
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
sns.histplot(df['Age'].dropna(), kde=True)
plt.title('Histogram of Age')
plt.subplot(1, 2, 2)
sns.boxplot(x=df['Age'].dropna())
plt.title('Boxplot of Age')
plt.tight_layout()
plt.show()
# Fare Histogram and Boxplot
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
sns.histplot(df['Fare'].dropna(), kde=True)
plt.title('Histogram of Fare')
plt.subplot(1, 2, 2)
sns.boxplot(x=df['Fare'].dropna())
plt.title('Boxplot of Fare')
plt.tight_layout()
plt.show()
# SibSp Histogram and Boxplot
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
sns.histplot(df['SibSp'].dropna(), kde=True)
plt.title('Histogram of SibSp')
plt.subplot(1, 2, 2)
sns.boxplot(x=df['SibSp'].dropna())
plt.title('Boxplot of SibSp')
plt.tight_layout()
plt.show()
# Parch Histogram and Boxplot
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
sns.histplot(df['Parch'].dropna(), kde=True)
plt.title('Histogram of Parch')
plt.subplot(1, 2, 2)
sns.boxplot(x=df['Parch'].dropna())
plt.title('Boxplot of Parch')
plt.tight_layout()
plt.show()
```

```python
# Correlation Matrix
corr = df[['Age', 'Fare', 'SibSp', 'Parch', 'Survived']].apply(pd.to_numeric, errors='coerce').corr()
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
# Pairplot with cleaned numeric data
cleaned_df = df[['Age', 'Fare', 'SibSp', 'Parch', 'Survived']].apply(pd.to_numeric, errors='coerce').dropna()
sns.pairplot(cleaned_df, hue='Survived')
plt.show()
```

### Basic Inference Tips

- Look for skewness in histograms to identify data distribution (e.g., Age is right-skewed).
- Use boxplots to detect outliers (e.g., Fare shows extreme values).
- Correlation matrix reveals linear relationships:
  - `SibSp` and `Parch` have weak correlations with survival.
  - `Fare` has a mild positive correlation with `Survived`.
- Pairplot helps visualize clusters or separations between survivors and non-survivors.
